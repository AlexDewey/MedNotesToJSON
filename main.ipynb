```python
from JSONcreate import get_json_full, get_json_segment
from JSONevalute import compare_json
import pandas as pd
import json
```

```python
# Get data
data = pd.read_csv('./data/train.csv')

num_samples = 100

results = 0
semi_results = 0
for index in range(0, num_samples):
    row = data.iloc[index]
    json_string = get_json_full(model='gpt-oss', notes=row['Note']) #gpt-oss #thewindmom/llama3-med42-8b

    ground_truth_obj = json.loads(row['json'])
    generated_obj = json.loads(json_string)

    print(ground_truth_obj)
    print(generated_obj)

    similarity = compare_json(ground_truth_obj, generated_obj)
    results += similarity
    print(f"Similarity with ground truth: {similarity:.2f}%\n")

print(f"Average similarity over {num_samples} samples: {results / num_samples:.5f}%")
```
